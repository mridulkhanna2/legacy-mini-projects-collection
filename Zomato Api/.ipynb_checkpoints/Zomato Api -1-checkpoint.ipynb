### Zomato Restaurant 2.2

Fetch the details of "Pa Pa Ya" restaurant using Zomato API.

Print the user rating, the average cost for two, cuisines and address of "Pa Pa Ya" restaurant.

Output Format:
user_rating
average_cost_for_two
cuisines
address

```python
## Open and read data file as specified in the question
## Print the required output in given format
import requests

p={'q':'Pa Pa Ya'}

r=requests.get("https://developers.zomato.com/api/v2.1/search",headers={"user-key":"87f15daab744d85971894165c35423cd"},params=p)

data=r.json()

for i in data['restaurants']:
    if i["restaurant"]["name"] == "Pa Pa Ya":
        print(i["restaurant"]['user_rating']['aggregate_rating'])
        print(str(i["restaurant"]['average_cost_for_two']))
        print((i["restaurant"]['cuisines']).strip())
        print((i["restaurant"]['location']['address']).strip())
```
